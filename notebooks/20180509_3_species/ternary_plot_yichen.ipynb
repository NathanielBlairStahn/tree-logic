```python
%matplotlib inline

import ternary
```

```python
import random

def random_points(num_points=25, scale=1):
    points = []
    for i in range(num_points):
        x = random.uniform(0, scale)
        y = random.uniform(0, scale - x)
        z = scale - x - y
        points.append((x,y,z))
    return points
```

```python
# Scatter Plot
scale = 1
figure, tax = ternary.figure(scale=scale)
figure.set_size_inches(10, 10)
tax.set_title("Scatter Plot", fontsize=20)
tax.boundary(linewidth=2.0)
tax.gridlines(multiple=0.1, color="blue")
points = random_points(5, scale=scale)
tax.scatter(points, marker='s', color='red', label="Red Squares")
points = random_points(10, scale=scale)
tax.scatter(points, marker='D', color='green', label="Green Diamonds")
points = random_points(15, scale=scale)
tax.scatter(points, marker='H', color='blue', label="Blue Diamonds")
tax.legend()
tax.ticks(axis='lbr', linewidth=1, multiple=0.1,tick_formats='%.1f')
tax.clear_matplotlib_ticks()

tax.show()
```
